# Deciding Columns to Keep Based on Their Description

| Field | Decision | Why |
|-------|----------|-----|
| **CREDIT_SCORE** | Keep | FICO score of the borrower at the time the loan was issued (300–850). This is the single most important indicator of borrower quality. |
| **FIRST_PYMT_DATE** | Keep | Date of the first mortgage payment in `YYYYMM` format. Required for quarterly analysis. |
| **FIRST_TIME_BUYER** | Keep | First-time buyers during the housing bubble were often less financially stable. |
| **MATURITY_DATE** | Drop | Redundant with **LOAN_TERM**. |
| **MSA** | Drop | Metropolitan Statistical Area code. Contains 455 unique values and 173,697 null values (~31% of the data). Geographic information is useful, but this is largely redundant because **PROPERTY_STATE** is available. |
| **MI_PCT** | Keep | Mortgage insurance percentage. This is the lender's own risk signal—if mortgage insurance was required, the lender already considered the loan riskier. |
| **NUM_UNITS** | Keep | Number of units in the property. Multi-unit properties are often investor purchases, and investors tend to walk away from loans more frequently than owner-occupants. |
| **OCC_STATUS** | Keep *(May Drop Later)* | Occupancy status (Primary Residence, Investor Property, Second Home). Since investor properties are generally riskier, this may be useful. It may later prove redundant with **NUM_UNITS**. |
| **OCLTV** | Keep | Original Combined Loan-to-Value ratio. Includes both the primary mortgage and any secondary loans. It contains more information than **LTV**. For example, a borrower could have **LTV = 80** but **OCLTV = 100** due to a second mortgage. Higher OCLTV indicates higher risk. |
| **DTI** | Keep | Debt-to-Income ratio. Total monthly debt payments divided by total monthly income at origination. Indicates how financially stretched the borrower was. |
| **UPB** | Keep | Unpaid Principal Balance. Larger loans generally carry greater absolute financial risk. |
| **LTV** | Keep | Loan-to-Value ratio. Although it contains less information than **OCLTV**, it is still a valuable measure of mortgage risk. |
| **INTEREST_RATE** | Keep | Interest rate at loan origination. Higher rates generally indicate that the lender priced in greater borrower risk. |
| **CHANNEL** | Keep | Indicates how the loan was originated. Broker-originated loans tended to be riskier because brokers often had little incentive to verify borrower income or assets thoroughly. |
| **PPM_FLAG** | Drop | Indicates whether the borrower is penalized for paying off the loan early. This is a very weak predictor since borrowers who default generally do not repay the loan anyway. |
| **PRODUCT_TYPE** | Drop | Indicates Fixed-Rate or Adjustable-Rate mortgages. The dataset contains only one unique value, resulting in zero variance and no predictive information. |
| **PROPERTY_STATE** | Keep | Two-letter state abbreviation representing the property's location. Geographic location is an important factor in mortgage performance. |
| **PROPERTY_TYPE** | Keep | Different property types exhibit different default rates. |
| **ZIP_3** | Drop | First three digits of the ZIP code. Contains 892 unique values, resulting in high cardinality. It is also largely redundant with **PROPERTY_STATE**. |
| **LOAN_SEQ_NUM** | Keep *(For Joining Only)* | Unique identifier for each loan. Used only for joining datasets. |
| **LOAN_PURPOSE** | Keep | Purpose of the loan (e.g., Purchase, Cash-out Refinance, etc.). |
| **LOAN_TERM** | Keep | Number of scheduled monthly payments. May contain non-standard values and could require transformation. |
| **NUM_BORROWERS** | Keep | Having a co-borrower provides an additional income source, potentially reducing default risk. |
| **SELLER_NAME** | Drop | Name of the entity that sold the mortgage to Freddie Mac. Not required for our analysis. |
| **SERVICER_NAME** | Drop | Name of the loan servicer as of the last reported period. Not required for our analysis. |
| **SUPER_CONF_FLAG** | Drop | Marks loans above conforming loan limits. Applicable only from October 2008 onward and contains nearly 99% null values. |
| **PRE_HARP_LOAN_SEQ** | Drop | Links loans refinanced under HARP. Since HARP was introduced in 2009, this field contains nearly 99% null values. |
| **SPECIAL_ELIGIBILITY** | Drop | Identifies loans under Freddie Mac special programs (e.g., Home Possible). Available only after March 2015. Contains zero variance, with all values marked as "Not Available." |
| **RELIEF_REFINANCE_FLAG** | Drop | Indicates HARP relief refinance loans. Since this is a post-2008 program, approximately 96% of the values are null. |
| **PROPERTY_VAL_METHOD** | Drop | Method used to value the property. Available only after 2017 and contains only one unique value (zero variance). |
| **INTEREST_ONLY_FLAG** | Drop | Indicates whether the borrower initially paid only interest without reducing principal. Contains only one unique value across the dataset. |
| **MI_CANCELLATION_FLAG** | Drop | Indicates whether mortgage insurance was cancelled after Freddie Mac purchased the loan. This is a post-origination event and has zero variance. |
| **ORIG_YEAR** | Keep | Year extracted by our `orig_loader()` function. Useful for time-based analysis. |

# Performance File — Columns We Are Loading

| Field | Decision | Why |
|-------|----------|-----|
| **LOAN_SEQ_NUM** | Keep | This is the unique loan identifier that links each performance record back to the origination data. Without this we cannot join the two files. |
| **DELINQ_STATUS** | Keep | This tracks how many months past due the borrower is in each monthly record. A value of 3 or more means the borrower is 90 or more days past due which is the standard definition of serious delinquency and what we are using as our default threshold. |
| **ZERO_BAL_CODE** | Keep | This tells us why the loan balance went to zero. We are only interested in two codes. Code 03 means the loan ended in a short sale or charge off where the lender accepted a loss. Code 09 means the bank foreclosed on the property, took ownership and sold it themselves which is the classic REO outcome. For most loans this field will be null because the loan never terminated during our observation window. |

```python
from src.loader import perf_loader

perf = perf_loader()

print("Shape:", perf.shape)
print("\nSample:")
print(perf.head())
print("\nColumn types:")
print(perf.dtypes)
```

As we can perf contains monthly data we neeed to handle it first and then merge 

```python
def collapse_perf(perf):
    """
    Collapse performance data from one row per loan per month
    to one row per loan.
    We take the worst value of each column across all months
    so we capture if a loan ever reached serious delinquency
    or ever got foreclosed at any point in its history.
    """
    perf_collapsed = perf.groupby("LOAN_SEQ_NUM").agg(
        MAX_DELINQ  = ("DELINQ_STATUS", "max"), # worst delinquency the loan ever reached
        ZERO_BAL    = ("ZERO_BAL_CODE", "max"), # foreclosure code if it ever happened
        PERF_MONTHS = ("PERF_YEAR", "count")    # total months of performance history
    ).reset_index()

    return perf_collapsed
```

# Transformer features as folllow
- DELINQ_STATUS->MAX_DELINQ # worst delinquency the loan ever reached
- ZERO_BAL_CODE->ZERO_BAL # foreclosure code if it ever happened
- PERF_YEAR->PERF_MONTHS # total months of performance history


```python
perf = collapse_perf(perf)
```

```python
df = df.merge(perf, on="LOAN_SEQ_NUM",how='left')
```
